## Reading data

```python
import pandas as pd 
from pathlib import Path

Data_PATH = Path("..") / "data" / "keywords" / "keyword_dataset_2026-06.csv"

df = pd.read_csv(Data_PATH)
print(f" (keyword_dataset_2026) Shape : {df.shape}")
print(f" (keyword_dataset_2026) Columns : {df.info()}")
```

How many unique itineraries are there

```python
print(f"Unique tours: {df['tour_url'].nunique():,}")
```

How many unique cities and which one are they 

```python
print(f"Cities tours: {df['city'].nunique():,}")
df['city'].value_counts().head(10)

```

```python
print(f"Missing cities before: {df['city'].isna().sum()}")
```

```python
df["city"] = df["city"].fillna("Unknown")
print(f"Missing cities after:  {df['city'].isna().sum()}")
```

> **Note on `fillna("Unknown")`:** this is for *exploration in this notebook only*. In the
> pipeline a blank `city` is intentional for generic days ("Day at Sea", "Farewell"), so
> don't fill at source.
>
> To **rename / merge** cities permanently (e.g. `Kimberley → Broome`), don't edit the
> DataFrame here — add a line to `data/config/city_synonyms.txt` (or fill the **"Merge →"**
> column in `city_review.xlsx` and run `python analysis/city_review_tool.py apply`), then
> re-run the pipeline. The rename then applies everywhere automatically.
